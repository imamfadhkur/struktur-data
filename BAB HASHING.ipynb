```python
BAB HASHING
    hashing adalah proses pengindeksan dan pengambilan elemen dalam struktur data untuk memberikan cara yang lebih cepat
dalam menemukan elemen menggunakan kunci hash. kunci hash ditentukan dengan cara data/nilai di modulus banyaknya kolom list. 
searching dalam hashing menggunakan hash key yang telah ditentukan.
di dalam hashing ada sebuah masalah yang dinamakan collision.
collision adalah ketika ada 2 item atau lebih memiliki slot yang sama, kita harus memiliki metode sistematis untuk
menempatkan item kedua atau selanjutnya di dalam tabel hash. Ada dua cara yaitu open addressing (linear dan quadratic)
dan close addressing (chaining probbing)
a. open addressing
    a) linear probbing
           Linear Probbing yaitu metode penempatan dengan cara melihat secara berurutan, slot demi slot,
        sampai menemukan sebuah posisi terbuka atau kosong. Jika ditemukan subuah posisi kosong maka item yang memiliki
        hash key yang sama ditaruh dalam posisi kosong tersebut. searching dalam linear probbing menggunakan cara linear.
```

```python
# data yang dimasukkan = 54,26,93,17,77,31,44,55,20

table = [None] * 11

def hash(x):
  return x % 11

def insert(table,key,value):
  index = hash(key)
  if table[index] == None:
    table[index] = value
  else:
    collusion=index
    found = False
    ind=collusion+1
    if ind>= len(table)-1:
      ind = 0
    while (ind<=len(table)-1) and not(found):
      if table[ind]== None:
        found=True
        table[ind]=value
      ind=ind+1

print(table)
insert(table,54,54)
print(table)
insert(table,26,26)
print(table)
insert(table,93,93)
print(table)
insert(table,17,17)
print(table)
insert(table,77,77)
print(table)
insert(table,31,31)
print(table)
insert(table,44,44)
print(table)
insert(table,55,55)
print(table)
insert(table,20,20)
print(table)
insert(table,25,25)
print(table)
```

```python
    b) Quadratic Probbing
            Quadratic probing adalah skema pengalamatan terbuka dalam pemrograman komputer untuk menyelesaikan collision
        dalam tabel hash, ketika nilai hash data yang masuk menunjukkan itu harus disimpan dalam slot atau
        bucket yang sudah diduduki. Quadratic probing beroperasi dengan mengambil indeks hash asli dan menambahkan nilai
        berturut-turut dari polinomial kuadratik sewenang-wenang sampai slot terbuka ditemukan.
        kuadrat probing yaitu metode penempatan yang berarti bahwa jika nilai hash pertama adalah h,
        nilai berturut-turut adalah h + 1, h + 4, h + 9, h + 16. begitu seterusnya.
        
```

```python
table = [None] * 10
def hash(x):
  return x % 10


def insert(table,key,value):
  index = hash(key)
  if table[index] == None:
    table[index] = value
  else:
    collusion=index
    found = False
    i=1
    ind = collusion+1

    if ind > len(table)-1:
        ind = 0

    while (ind<=len(table)-1) and not(found):
      if table[ind]== None:
        found=True
        table[ind]=value
        ind = (collusion+(i*i)) % 11
      i=i+1

print(table)
insert(table,54,54)
print(table)
insert(table,26,26)
print(table)
insert(table,93,'93')
print(table)
insert(table,17,'17')
print(table)
insert(table,77,'77')
print(table)
insert(table,31,'31')
print(table)
insert(table,44,'44')
print(table)
insert(table,55,'55')
print(table)
```

```python
b. Close Addressing(Chaining)
        Chaining adalah metode yang memungkinkan banyak item terdapat pada lokasi yang sama di tabel hash.
    Ketika collision terjadi, item tersebut masih ditempatkan di slot yang tepat dari tabel hash. Seperti cotoh data dibawah ini
    [[77, 44, 55], [], [], [], [26], [93], [17], [], [], [31, 20], [54]]
```
